# Toplogical Sort

A topological ordering of a directed graph is that for every directed edge uv from vertx u to vertex v, u comes before v.


Definition: A topological order of a directed graph G is a labeling f of G's nodes such that:
1. the f(v)'s are the set {1,2,...n}
1. (u,v) $\in G \implies f(u) < f(v) $

# Derivative Definition

The derivative is defined as the following equation:

$$
\begin{equation*}
f'(x) = \frac{df(x)}{dx} = lim_{h -> 0} \, \ \frac{f(x+h) - f(x)}{h}
\end{equation*}
$$

----

The gradient is the tangent line to a function at a point given that h goes to zero in the limit. You can think of this as the slope of the two points in interactive plot.

The derivative has a unique property such that the derivative always points in the direction of the steepest ascent.

Using this property and treating the derivative as a vector, or using the properties of graphs, the derivative can be scaled by -1 or reflected to point in the opposite direction. As a result, this causes the derivative to point towards the steepest descent.

```python
# interactive code here to create a graph and plot the line between two points, making a triangle
from ipywidgets import interact, interactive, IntSlider, Layout, interact_manual
from ipywidgets import *
import ipywidgets as widgets
from IPython.display import display, clear_output

import matplotlib.pyplot as plt
from matplotlib.pyplot import Figure
import numpy as np

%matplotlib notebook
```

```python
v = np.arange(-10,10,.01)

x1 = -5.0
x2 = 5.0

def w(x):
    return (x-2)*(x+2)*(x+1)

def diag(x1,x2):
    y1 = w(x1)
    y2 = w(x2)    
    m = (y2-y1)/(x2-x1)
    return (lambda x : m*(x - x1) + y1)

fig,axe = plt.subplots(1,1)

#f
function = axe.plot(v,w(v))

#diagonal line
line, = axe.plot(v,diag(x1,x2)(v) , color='g')

#horizontal
horz, = axe.plot(v,v*0+w(x1), color='g')

point1, = axe.plot(x1,w(x1), marker='o', color='r')

point2, = axe.plot(x2,w(x2), marker='o', color='r')

#vertical
vert = axe.axvline(x2, color='g')

print(vert)

def f(x):
    x1,x2 = x.new[0], x.new[1]
    
    line.set_data(v,diag(x1,x2)(v))
    horz.set_data(v,v*0+w(x1))
        
    point1.set_data(x1,w(x1))
    point2.set_data(x2,w(x2))
    
    
    del axe.lines[-1]
    axe.axvline(x2, color='g')
    
    fig.canvas.draw()

def make_slider():
    slider = widgets.FloatRangeSlider(value=[x1,x2], min=-10, max=10)
    slider.observe(f,names='value')
    display(slider)


make_slider()
plt.show()
```

# The Partial Derivative

Similarly, the two dimensional case generalizes to the n-dimensional case for the multivariable function. For a multivariable function, the partial derivative is defined as the following equation:

$$
\begin{equation*}
\nabla_{x_1} = \frac{\partial f(x_1, x_2, \cdots , x_n)}{\partial x_1} = lim_{h -> 0} \, \ \frac{f(x_1+h, x_2, \cdots , x_n) - f(x_1, x_2, \cdots , x_n)}{h}
\end{equation*}
$$

For the three dimensional case, imagine having a surface, and then slicing the surface along one of the input dimensions. Then take two points. The first point at x and the second point a position at x+h. Now evaluate the rise over run or the slope. This would be the derivative. A partial derivative is letting other variables be constants or imagining a slice of that variable in space.

Below is an example of an eliptical function and their respective partial derivatives.

$$ z = f(x,y) = x^2 + xy + y^2 $$

$$ \frac{\partial z}{\partial x} = \frac{\partial f(x,y)}{\partial x} = 2x + y $$

$$ \frac{\partial z}{\partial y} = \frac{\partial f(x,y)}{\partial y} = 2y + x $$

# The Gradient

The gradient is a collection of partial derivatives for every variable in multivariable function.

$$
\nabla f(x_1, x_2, \cdots, x_n) =
\begin{bmatrix}
\frac{ \partial f(x_1, x_2, \cdots, x_n) }{ \partial x_1} \\
\frac{ \partial f(x_1, x_2, \cdots, x_n) }{ \partial x_2} \\
\vdots \\
\frac{ \partial f(x_1, x_2, \cdots, x_n) }{ \partial x_n}
\end{bmatrix}
$$

The gradient above is a vector of partial derivative with respect to each of its principal axis $ x_1 \cdots x_n $


# The Numerical Gradient

The numerical gradient is the calculation of the gradient using the derivative definition.

So in practice, the following is computed programatically to find the derivative. We select a value for h and evalate the function at the original location and the pertubated location.

$  
\nabla f(x,y) = 
\begin{bmatrix} 
\frac{df(x,y)}{dx} \\
\frac{df(x,y)}{dy}
\end{bmatrix} = 
\begin{bmatrix}
\frac{f(x+h, y) - f(x, y)}{.00001} \\
\frac{f(x, y+h) - f(x, y)}{.00001}
\end{bmatrix}
$

The numerical gradient is useful in validating the analytical gradient. This is usually used to debug code. The numerical gradient is not used in practice.

# The Analytical Gradient

The analytical gradient is the results of the evaluation of the gradient.

for ex:

$ y = x^2 \text{ then the analytical gradient is } \frac{dy}{dx} = 2x  $

The analtical gradient is used in practice as it very perfomant.

# Automatic Differentiation

Automatic differentiation is a way to automatically calculate the gradient of a function for its arguments with respect to other variables. This takes advantage of the fact that most equations can be represented as a DAG. A DAG can be topoloically sorted. After the nodes are sorted, you can traverse the graph to calculate all the gradients of a function. This is basically the chain rule.

Lets look at a single node:

We can calculate the local gradient of a function immediately during the forward pass.
The local gradients ares $ \frac{dz}{dx} and \frac{dz}{dy}$
However, we cannot yet calculate $ \frac{dL}{dx} and \frac{dL}{dy}$
We are waiting for the z function to calculate its derivative with respect to final equation.



Once we have a derivative, we can then use the gloabl derivative and the local derivative to compute $ \frac{dL}{dx} and \frac{dL}{dy}$

This is ultimately just the chain rule.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-10,11,.1)
y = np.arange(-10,11,.1)

X,Y = np.meshgrid(x,y)

Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X,Y,Z)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.show()
```

# Activation Functions

All activation functions are element wise functions. This means that doing this operation on a vector/matrix will actually apply the operation at each element in the vector/matrix.

## Sigmoid

Given: 
$ x \in \mathbb R^n  $

$ \sigma(x) = \sigma([x_1, x_2, ...]) = [\sigma(x_1), \sigma(x_2), ... , \sigma(x_n)] $

$ \sigma(x) : \mathbb R^n \rightarrow \mathbb (0,1)^n $

$$ \sigma(x) = \frac{1}{1+e^{-x}} $$



```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.arange(-10,10,.01)
def sigmoid(x):
    return 1. / (1. + np.exp(-x))

plt.plot(x,sigmoid(x))
plt.title("sigmoid")
```

## tanh

Given: 
$ x \in \mathbb R^n  $

$ tanh(x) : \mathbb R^n \rightarrow \mathbb (-1,1)^n $

$$ tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} = 2\sigma(2x)+1 $$




```python
x = np.arange(-10,10,.01)
def tanh(x):
    return (2 * sigmoid(2*x) - 1)

plt.plot(x,tanh(x))
plt.title("tanh")
```

# relu stands for Rectified linear unit

Relu was a byproduct of softplus. This is by far the most effective activation function.

Given: 
$ x \in \mathbb R^n  $

$ relu(x) : \mathbb R^n \rightarrow \mathbb [0,\infty)^n $

$$
relu(x) = \left\{
        \begin{array}{ll}
            x & x \gt 0 \\
            0 & otherwise
        \end{array}
    \right.
$$

$$ relu(x) = max(x,0) $$



```python
x = np.arange(-10,10,.01)
def relu(x):
    return np.maximum(x,0)

plt.plot(x,relu(x))
plt.title("relu")
```

## Logistic Regression as a NN


$ v = h(x,w,b) = w^Tx + b $

$ z = \sigma(v) = \frac{1}{1+e^{-v}} $

$ L_i(z,y) = -y \, ln (z) - (1-y) \, ln (1-z) $

where, 

$ w \in \mathbb R^n $
w ~ N(0,1)

$ x \in \mathbb R^n $

$ b \in \mathbb R $ b ~ N(0,0)

$ v \in \mathbb R $

$ z \in \mathbb R $

$ y \in \{0,1\} $

$ L_i \in \mathbb R $

```python
def hypothesis(x,w,b):
    """
    x is a vector
    w is a vector
    b is a scalar
    """
    return np.dot(x,w) + b

def sigmoid(v):
    """
    v is a scalar
    """
    return 1./(1.+np.exp(-v))

def logistic_loss(z,y):
    """
    z is a vector
    y is a scalar
    """
    return -y * np.exp(z) - (1.-y) * np.exp(1-z)
    
def hypothesis_gradient(x,w,b,dL):
    dx = None
    dw = None
    db = None
    return dx,dw,db
    

def sigmoid_gradient(v,dL):
    dv = None
    return dv
    
def logistic_loss_gradient(z,y):
    dz = None
    dy = None
    return dz,dy

def train(x,y,w,b):
    v = hypothesis(x,w,b)
    t = sigmoid(v)
    loss = logistic_loss(t)
    dL,_ = logistic_loss_gradient(t)
    dt = sigmoid_gradient(v,dL)
    dx,dw,db = hypothesis_gradient(x,w,b,dt)
    return loss, dw, db;

def gradient_descent(x,y,alpha,epochs=1000):
    w = None
    b = None
    for _ in range(epochs):
        w = w - alpha * train(x,y,w,b)[1]
        b = b - alpha * train(x,y,w,b)[2]
    
```
